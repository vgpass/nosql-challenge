## Notebook Set Up

```python
# Import dependencies
from pymongo import MongoClient
from pprint import pprint
import pandas as pd
```

```python
# Create an instance of MongoClient
mongo = MongoClient(port=27017)
```

```python
# assign the uk_food database to a variable name
db = mongo['uk_food']
```

```python
# review the collections in our database
print(db.list_collection_names())
```

```python
# assign the collection to a variable
establishments = db['establishments']
```

### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {'scores.Hygiene': {'$eq': 20}}
twenty = establishments.find(query)

# Use count_documents to display the number of documents in the result
print("Number of documents in result:", establishments.count_documents(query))

# Display the first document in the results using pprint
pprint(twenty[0])
```

```python
# Convert the result to a Pandas DataFrame
twenty_df = pd.DataFrame(twenty)

# Display the number of rows in the DataFrame
print(f'The number of rows in the dataframe is {len(twenty_df)}')

# Display the first 10 rows of the DataFrame
twenty_df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query = {'RatingValue': {'$eq': '4'},
            'RatingValue': {'$eq': '5'},
            'LocalAuthorityName': {'$regex': 'London'}
         }

london_gte_4 = establishments.find(query)

# Use count_documents to display the number of documents in the result
print("Number of documents in result:", establishments.count_documents(query))

# Display the first document in the results using pprint
pprint(london_gte_4[0])
```

```python
# Convert the result to a Pandas DataFrame
london_gte_4_df = pd.DataFrame(london_gte_4)

# Display the number of rows in the DataFrame
print(f'The number of restaurants in the dataframe is {len(london_gte_4_df)}')

# Display the first 10 rows of the DataFrame
london_gte_4_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of '5', sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01
latitude = 51.49014200
longitude = 0.08384000

lat_plus = latitude + degree_search
lat_minus = latitude - degree_search
\
lng_plus = longitude + degree_search
lng_minus = longitude - degree_search

query = {'RatingValue': {'$eq': '5'},
        'geocode.latitude': {'$gte': lat_minus},
        'geocode.latitude': {'$lte': lat_plus},
        'geocode.longitude': {'$gte': lng_minus},
        'geocode.longitude': {'$lte': lng_plus},
}
sort =  [('scores.Hygiene', 1)]

nearby_locations = establishments.find(query).sort(sort)

# Print the results
print("Number of documents in result:", establishments.count_documents(query))

# for location in nearby_locations:
#     pprint(location)
```

```python
# Convert result to Pandas DataFrame
nearby_locations_df = pd.DataFrame(nearby_locations)
print(f'The number of restaurants in the dataframe is {len(nearby_locations_df)}')
nearby_locations_df.head(15)
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
match_query = {'$match': {'scores.Hygiene': {'$eq': 0}}}

# 2. Groups the matches by Local Authority
group_query = {'$group': {'_id': '$LocalAuthorityName', 'count': { '$sum': 1 }}}

# 3. Sorts the matches from highest to lowest
sort_values = {'$sort': { 'count': -1}}

pipeline = [match_query, group_query, sort_values]

results = list(establishments.aggregate(pipeline))

# Print the number of documents in the result
# print("Number of documents in result:", establishments.count_documents(query))
print("Number of restaurants in result: ", len(results))

# Print the first 10 results
for result in range(10):
    print(results[result])
```

```python
# Convert the result to a Pandas DataFrame
zero_hygiene_df = pd.DataFrame(results)

# Display the number of rows in the DataFrame
print(f'The number of restaurants in the dataframe is {len(zero_hygiene_df)}')

# Display the first 10 rows of the DataFrame
zero_hygiene_df.head(10)
```
